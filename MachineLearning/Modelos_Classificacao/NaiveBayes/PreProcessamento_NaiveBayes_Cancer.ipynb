### Pré-Processamento

#### Diagnóstico(diagnosis) --> (M = maligno, B = benigno)

*   Maligno = Acontecem quando uma célula do organismo passa a se replicar de maneira desordenada e descontrolada;
*   Beligno = São constituídos por células bem semelhantes às que os originaram e não possuem a capacidade de replicação. 

```python
df_cancer['diagnosis'].value_counts()
```

```python
sns.countplot(x=df_cancer['diagnosis']);
```
